```python
import hanlp
import tqdm
import re
import pickle as pkl
recognizer = hanlp.load(hanlp.pretrained.ner.MSRA_NER_BERT_BASE_ZH)
```

```python
res = []
a = 0
lines = []
for line in open(r'../1_data/tian_long_ba_bu.txt', 'r', encoding='gb2312', errors='ignore'):
    line = line.strip()
    if len(line) >= 126:
        line = re.split('。|：|！|？|，|、| |,', line)
        for l in line:
            if len(l) >= 126:
                print(l)
            else:
                lines.append(l)
    else:
        lines.append(line)
all_data = []
for l in lines:
    all_data.append(list(l))
re = recognizer(all_data[:])
pkl.dump(re, open('1_hanlp_ner.pkl', 'wb'))
```
